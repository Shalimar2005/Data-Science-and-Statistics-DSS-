```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

x

```python
y=x*2
```

```python
y
```

Data visualization 

```python
plt.plot(x,y)
plt.title("line chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

Bar chart

```python
plt.bar(x,y,)
plt.title("bar chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.bar(x,y,color='green')
plt.title("bar chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

Scatter chart

```python
a=(10,20,30,40,50,60,70,80,90)
b=(11,22,33,44,55,66,77,88,99)

plt.scatter(a,b,color='black')
plt.title("Scatter chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
